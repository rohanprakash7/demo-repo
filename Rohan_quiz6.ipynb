```python
#1

import numpy as np
import matplotlib.pyplot as plt
population = np.random.randint(1, 100, size=10000)
sample_means = [np.mean(np.random.choice(population, size=30, replace=True)) for _ in range(1000)]
plt.figure(figsize=(7, 4))
plt.hist(sample_means, bins=30, color='red', edgecolor='brown', alpha=0.7)
plt.title('Distribution of 1,000 Sample Means (Sample Size = 30)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
population_mean = np.mean(population)
population_std = np.std(population)
sem = population_std / np.sqrt(30)#
print(f'Population Mean: {population_mean:.2f}')
print(f'Population Standard Deviation: {population_std:.2f}')
print(f'Standard Error of the Mean (SEM): {sem:.2f}')
```

```python
#1
# The code indicates that the distribution of the 1,000 sample means is approximately normal, with the population mean and mean and the standard deviation and mean being near to one another.
# This is in line with the Central Limit Theorem, which states that sample means will be roughly normal even when the population distribution is non-normal.
```

```python
#2
#With a mean that is comparable to the population mean and a standard deviation that is comparable to the standard error of the mean, the code suggests that the 500 sample means have a distribution that is approximately normal.
#This is consistent with the Central Limit Theorem, which asserts that a normal distribution exists even in non-normal population distributions.
```

```python
#3
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
np.random.seed(0)
population = np.random.exponential(scale=2, size=10000)
sample_means = np.zeros(500)
for i in range(500):
    sample = np.random.choice(population, size=50, replace=False)
    sample_means[i] = np.mean(sample)
population_mean = np.mean(population)
population_std = np.std(population)
sem = population_std / np.sqrt(50)
confidence_interval = stats.t.interval(confidence=0.95, df=len(sample_means)-1, loc=np.mean(sample_means), scale=stats.sem(sample_means))
print(f'Population Mean: {population_mean:.2f}')
print(f'Population Standard Deviation: {population_std:.2f}')
print(f'Standard Error of the Mean (SEM): {sem:.2f}')
print(f'95% Confidence Interval: ({confidence_interval[0]:.2f}, {confidence_interval[1]:.2f})')
if confidence_interval[0] < population_mean < confidence_interval[1]:
    print('The Population mean is within the 95% confidence interval.')
else:
    print('The Population mean is not within the 95% confidence interval.')
```

```python
#3
#INTERPRETATION
#The 95% confidence interval, such as 1.93, 2.07, represents the 95% probability that the true population mean falls within that interval.
#The sample mean, not the population mean itself, appears to be a trustworthy approximation of the population mean if the population mean falls within this range.
```

```python
#4
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
population = np.random.randint(1, 101, size=10000)
sample_means_10 = np.zeros(1000)
sample_means_30 = np.zeros(1000)
for i in range(1000):
    sample_10 = np.random.choice(population, size=10, replace=False)
    sample_means_10[i] = np.mean(sample_10)

    sample_30 = np.random.choice(population, size=30, replace=False)
    sample_means_30[i] = np.mean(sample_30)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(sample_means_10, bins=30, density=True, alpha=0.7, color='brown', edgecolor='black')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Sample Size = 10')
plt.subplot(1, 2, 2)
plt.hist(sample_means_30, bins=30, density=True, alpha=0.7, color='brown', edgecolor='black')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')
plt.title('Sample Size = 30')
plt.tight_layout()
plt.show()
std_10 = np.std(sample_means_10)
std_30 = np.std(sample_means_30)
print(f'Standard Deviation of Sample Means (Sample Size = 10): {std_10:.2f}')
print(f'Standard Deviation of Sample Means (Sample Size = 30): {std_30:.2f}')
```

```python
#4
#Given that smaller sample sizes lead to bigger standard errors of the mean (SEM), the code demonstrates that a sample size of 10 produces a more dispersed distribution of sample means. Reduced standard deviation and decreased variability are indicative of larger sample sizes and more accurate population mean estimates.
#A normal distribution is shown by this, which is consistent with the Central Limit Theorem.
```

Problem 1 - T-test

```python
#Problem1:

import numpy as np
from scipy import stats
class_A_scores = [85, 90, 78, 92, 88, 76, 95, 87, 79, 91]
class_B_scores = [80, 82, 88, 85, 83, 87, 84, 86, 89, 81]
t_stat, p_val = stats.ttest_ind(class_A_scores, class_B_scores)
print(f'T-statistic: {t_stat:.2f}')
print(f'p-value: {p_val:.4f}')
if p_val < 0.05:
    print('There is a significant difference in the mean scores between the two classes.')
else:
    print('There is no significant difference in the mean scores between the two classes.')
```

Problem 2 - Chi-Squared test


```python
#Problem2:

import numpy as np
from scipy import stats
observed_frequencies = np.array([[30, 10, 10], [10, 30, 10]])
chi2, p_val, dof, expected = stats.chi2_contingency(observed_frequencies)
print(f'Chi-squared statistic: {chi2:.2f}')
print(f'Degrees of freedom: {dof}')
print(f'p-value: {p_val:.4f}')
if p_val < 0.05:
    print('There is a significant association between grade level and beverage preference.')
else:
    print('There is no significant association between grade level and beverage preference.')
```

Problem 3 - One-way ANOVA

```python
#Problem3:

import numpy as np
from scipy import stats
method_1_scores = [85, 87, 88, 86, 84, 85, 87]
method_2_scores = [80, 82, 81, 83, 82, 80, 81]
method_3_scores = [90, 91, 92, 90, 91, 92, 93]
f_stat, p_val = stats.f_oneway(method_1_scores, method_2_scores, method_3_scores)
print(f'F-statistic: {f_stat:.2f}')
print(f'p-value: {p_val:.4f}')
if p_val < 0.05:
    print('There is a significant difference in the mean scores between the teaching methods.')
else:
    print('There is no significant difference in the mean scores between the teaching methods.')
```

Problem 4 One sample Z-test

```python
#Problem4:
import numpy as np
from scipy import stats
sample_scores = [78, 76, 74, 75, 77, 76, 78, 74, 79, 75]
population_mean = 75
population_std = 10
sample_mean = np.mean(sample_scores)
std_err = population_std / np.sqrt(len(sample_scores))
z_score = (sample_mean - population_mean) / std_err
p_val = stats.norm.sf(abs(z_score))
print(f'Sample mean: {sample_mean:.2f}')
print(f'Z-score: {z_score:.2f}')
print(f'p-value: {p_val:.4f}')
if p_val < 0.05:
    print('There is a significant difference between the sample mean and the population mean.')
else:
    print('There is no significant difference between the sample mean and the population mean.')
```

Problem 5 Two Sample Z-Test

```python
#Problem5:
import numpy as np
from scipy import stats
class_X_scores = [85, 87, 88, 86, 84, 85, 87]
class_Y_scores = [80, 82, 81, 83, 82, 80, 81]
population_variance_X = 15
population_variance_Y = 20
sample_mean_X = np.mean(class_X_scores)
sample_mean_Y = np.mean(class_Y_scores)
std_err_X = np.sqrt(population_variance_X / len(class_X_scores))
std_err_Y = np.sqrt(population_variance_Y / len(class_Y_scores))
z_score = (sample_mean_X - sample_mean_Y) / np.sqrt(std_err_X**2 + std_err_Y**2)
p_val = stats.norm.sf(abs(z_score))
print(f'Sample mean X: {sample_mean_X:.2f}')
print(f'Sample mean Y: {sample_mean_Y:.2f}')
print(f'Z-score: {z_score:.2f}')
print(f'p-value: {p_val:.4f}')
if p_val < 0.05:
    print('There is a significant difference between the mean scores of the two classes.')
else:
    print('There is no significant difference between the mean scores of the two classes.')
```

```python
#6
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

data = pd.DataFrame({
    'Student ID': range(1, 11),
    'Group': ['A']*5 + ['B']*5,
    'Score': [85, 87, 82, 90, 88, 78, 80, 79, 83, 81]
})
```

```python
#task 1
plt.figure(figsize=(10, 6))
plt.subplot(1, 2, 1)
plt.hist(data[data['Group'] == 'A']['Score'], bins=5, alpha=0.5, label='Group A')
plt.hist(data[data['Group'] == 'B']['Score'], bins=5, alpha=0.5, label='Group B')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.title('Distribution of Scores')
plt.legend()
plt.subplot(1, 2, 2)
plt.boxplot([data[data['Group'] == 'A']['Score'], data[data['Group'] == 'B']['Score']], labels=['Group A', 'Group B'])
plt.xlabel('Group')
plt.ylabel('Score')
plt.title('Boxplot of Scores')
plt.show()
# Observations from the visualization:
# The distribution of scores for Group A is slightly skewed to the right, while Group B is more symmetric.
# The boxplot shows that Group A has a higher median score and a larger interquartile range (IQR) compared to Group B.
```

```python
#task 2
group_A_mean = data[data['Group'] == 'A']['Score'].mean()
group_A_std = data[data['Group'] == 'A']['Score'].std()
group_B_mean = data[data['Group'] == 'B']['Score'].mean()
group_B_std = data[data['Group'] == 'B']['Score'].std()
print(f'Group A: Mean = {group_A_mean:.2f}, Std Dev = {group_A_std:.2f}')
print(f'Group B: Mean = {group_B_mean:.2f}, Std Dev = {group_B_std:.2f}')
```

```python
#task 3
t_stat, p_val = stats.ttest_ind(data[data['Group'] == 'A']['Score'], data[data['Group'] == 'B']['Score'])
print(f't-statistic: {t_stat:.2f}')
print(f'p-value: {p_val:.4f}')
if p_val < 0.05:
    print('There is a significant difference in the scores between Group A and Group B.')
else:
    print('There is no significant difference in the scores between Group A and Group B.')
```

```python
#task 4
if p_val < 0.05:
    print('The teaching method used for Group A is more effective than the teaching method used for Group B.')
```

```python
#task 5
if p_val < 0.05:
    print('The teaching method used for Group A is more effective because it resulted in higher mean scores and a larger range of scores, indicating a more comprehensive understanding of the material.')
else:
    print('There is no significant difference in the effectiveness of the teaching methods used for Group A and Group B.')
```

```python
#7
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
data_sleep = pd.DataFrame({
    'Day': range(1, 8),
    'Control': [7.5, 7.8, 7.6, 7.7, 7.9, 7.8, 7.6],
    'Treatment': [8.2, 8.4, 8.1, 8.3, 8.5, 8.2, 8.4]
})
```

```python
#task 1
plt.figure(figsize=(10, 6))
plt.plot(data_sleep['Day'], data_sleep['Control'], label='Control')
plt.plot(data_sleep['Day'], data_sleep['Treatment'], label='Treatment')
plt.xlabel('Day')
plt.ylabel('Sleep Duration (hours)')
plt.title('Average Sleep Duration Over the Week')
plt.legend()
plt.show()
# Observations from the visualization:
# The treatment group has a consistently higher average sleep duration than the control group over the week.
```

```python
#task 2
control_mean = data_sleep['Control'].mean()
treatment_mean = data_sleep['Treatment'].mean()
print(f'Control Group: Mean Sleep Duration = {control_mean:.2f} hours')
print(f'Treatment Group: Mean Sleep Duration = {treatment_mean:.2f} hours')
```

```python
#task 3
t_stat, p_val = stats.ttest_rel(data_sleep['Control'], data_sleep['Treatment'])
print(f't-statistic: {t_stat:.2f}')
print(f'p-value: {p_val:.4f}')
if p_val < 0.05:
    print('There is a significant difference in the sleep duration between the control and treatment groups.')
else:
    print('There is no significant difference in the sleep duration between the control and treatment groups.')
```
